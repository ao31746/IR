```python
#single link

import numpy as np

#print (up/(np.linalg.norm(a)*np.linalg.norm(b)))
def Sim_sigle(cos_vec1, cos_vec2):
    
    
    vc_norm_x = [i[1] for i in cos_vec1]
    vc_norm_y = [i[1] for i in cos_vec2]

    up = 0
    for i in cos_vec2:
        #print ([y[0] for x, y in enumerate(cos_vec[0]) if y[0] == i[0]])
        for j in cos_vec1:
            if i[0]==j[0]:
                up+=i[1]*j[1]
                
    down = np.linalg.norm(vc_norm_x) * np.linalg.norm(vc_norm_y)
    if down ==0:
        return 0
    else:
        return up/down

#Sim_sigle(5,7)
```
